```python

unit_cost=df['unit_cost']
qc_pass_rate=df['qc_pass_rate']
customer_rating_avg=df['customer_rating_avg']
damaged_on_arrival=df['damaged_on_arrival']
return_rate=df['return_rate']

cost_score = 1 - (unit_cost - min(unit_cost)) / (max(unit_cost) - min(unit_cost))


quality_score = (
    0.5 * qc_pass_rate +
    0.3 * (customer_rating_avg / 5.0) +
    0.2 * (1 - damaged_on_arrival)
)


supplier_score = (
    0.5 * quality_score +       
    0.3 * (1 - return_rate) +   
    0.2 * cost_score            
)*100
df['supplier_score']=supplier_score
```
